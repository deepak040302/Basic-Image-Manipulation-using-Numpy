```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
import skimage.io
```

```python
class ImageEditor:
    def __init__(self):
        
        
        print("---------------------------------------------------------")
        print("-----------Welcome to ABC Image Editing software---------")
        print("---------------------------------------------------------")
        
        self.__image_path=""  # variable to store the path of the inputed image
        self.__image=""      # variable to store the image
        self.__new_img="" 
        self.menu()  #calling the menu method to display the menu
        
    def menu(self):
        
        while True:
            print("""
                Menu:
                  1. Load image
                  2. Edit image
                  3. Save image
                  4. Print image
                  5. Quit
            """)
            print()
            option = input("Enter 1, 2, 3, 4 or 5: ")  #input option 
            print()
            
            if option not in {'1','2','3','4','5'}:    #checking that input option is valid or not
                print("Invalid Choice!! \n Please Enter : 1, 2, 3, 4 or 5 only!\n")
                print()
                continue
            else:
                if option == '1':            # if 1 select then calling load_image method
                    self.load_image()
                elif option == '2':          # if 2 select then calling edit_image method
                    self.edit_image()
                elif option == '3':          # if 3 select then calling save_image method
                    self.save_image()
                elif option == '4':          # if 4 select then calling quit method
                    self.print_image(self.__image) 
                elif option == '5':
                    self.quit()
                    break
   
    def edit_img_menu(self):  #edit image menu method
        
        while True:
            print("""    
                **********************************
                      Edit Image Menu
                **********************************
                    1. Horizontal Flip
                    2. Vertical Flip
                    3. Invert Colors
                    4. Convert to Grayscale
                    5. Convert to B&W
                    6. Go To Main Menu
                **********************************
            """)
            print()
            option = input("Enter 1, 2, 3, 4 ,5 or 6: ")  #input option 
            print()
            
            if option not in {'1','2','3','4','5','6'}:    #checking that input option is valid or not
                print("Invalid Choice!! \n Please Enter : 1, 2, 3, 4, 5 or 6 only!\n")
                print()
                continue
            else:
                if option == '1':            # if 1 select then horizontal_flip method
                    self.horizontal_flip()
                elif option == '2':          # if 2 select then vertical_flip method
                    self.vertical_flip()
                elif option == '3':          # if 3 select then calling invert_colors method
                    self.invert_colors()
                elif option == '4':          # if 4 select then calling convert_to_grayscale method
                    self.convert_to_grayscale() 
                elif option == '5':          # if 5 select then calling convert_to_BW method
                    self.convert_to_BW()
                elif option == '6':
                    break
                    
    def horizontal_flip(self):  # horizontal_flip method to flip the image to horizontal
        
        hor_flip_img=np.fliplr(self.__image) # flipping the image to horizontal
        self.__new_img=hor_flip_img  
        
        print("-----------------Performing Horizontal Flip-----------------")
        
        plt.subplot(121), plt.imshow(self.__image)
        plt.title('Original Image')   # printing the original image

        plt.subplot(122), plt.imshow(hor_flip_img)
        plt.title('Flipped Image')   # printing the horizontal flipped image
        
        plt.show()
        
    def vertical_flip(self):  # vertical_flip method to flip the image to vertical
        
        ver_flip_img=np.flipud(self.__image) # flipping the image to vertical
        self.__new_img=ver_flip_img
        
        print("-----------------Performing Vertical Flip------------------")
        
        plt.subplot(121), plt.imshow(self.__image)
        plt.title('Original Image')  # printing the original image

        plt.subplot(122), plt.imshow(ver_flip_img)
        plt.title('Flipped Image')  # printing the vertical flipped image
        
        plt.show()
        
    def invert_colors(self):  # invert_colors method to invert the colors of the image
        
        invert_col_img = skimage.color.rgb2hsv(self.__image) #inverting the color of the image
        self.__new_img=invert_col_img
        
        print("-----------------Inverting The color of the Image-----------------")
        
        plt.subplot(121), plt.imshow(self.__image)
        plt.title('Original Image')   # printing the original image

        plt.subplot(122), plt.imshow(invert_col_img)
        plt.title('Inverted Color Image')   # printing the inverted color image 

        plt.show()
    
    def convert_to_grayscale(self):  # horizontal_flip method to convert the image to graysacle
        
        org_img=skimage.io.imread(self.__image_path)
        gray_converted_img=skimage.io.imread(self.__image_path,as_gray=True)  #converting the image to grayscale format
        self.__new_img=gray_converted_img
        
        print("-----------------Converting The Image to Grayscale-----------------")
        
        plt.subplot(121), plt.imshow(org_img)
        plt.title('Original Image')  # printing the original image

        plt.subplot(122), plt.imshow(gray_converted_img)
        plt.title('Grayscale Image')  # printing the Graysacle image
        
        plt.show()
    
    def convert_to_BW(self):  # convert_to_bw method to convert the image to black and white
        
        black_and_white_img= skimage.color.rgb2gray(self.__image) # converting the image to black and white
        self.__new_img=black_and_white_img
        
        print("---------------Converting The Image to Black and white--------------")
        
        fig, axes = plt.subplots(1, 2, figsize=(8, 4))
        ax = axes.ravel()

        ax[0].imshow(self.__image)
        ax[0].set_title("Original Image")  # printing the original image
        
        ax[1].imshow(black_and_white_img, cmap=plt.cm.gray)
        ax[1].set_title("B&W Image")  # printing the converted black and white image

        fig.tight_layout()
        plt.show()
    
    def load_image(self):  # load_image method to load the image for further editing
        
        img=str(input("Please Enter the Path of the image : ")) #taking path of image as input 
        img=img.replace("\\","/")
        
        self.__image_path = img
        self.__image = io.imread(self.__image_path)
        
    def print_image(self,out):  # print_image method to print the current image
        
        print("-----------------Printing The Image------------------")
        plt.subplot(121), plt.imshow(out)  #printing the image
        plt.title('Original Image') 
        plt.show()
        
    def edit_image(self):  # edit_image method for to call the edit_image_menu
        self.edit_img_menu()
                        
    def save_image(self):  # save_image method for saving the image
        self.__image=self.__new_img
        print("Image Saved Succesfully...")
    
    def quit(self):  # quit method
        print("----------------------------------------")
        print("Thank You for using our editor services.")
        print("----------------------------------------")
        return
```
